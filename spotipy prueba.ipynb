```python
for e in audio_features:
    for d in e:
        df = pd.DataFrame([key for key in d.keys()], columns=['Name'])
        print(df)
        df['danceability'] = [value['danceability'] for value in d.values()]
        df['energy'] = [value['energy'] for value in d.values()]
        df['key'] = [value['key'] for value in d.values()]
        df['loudness'] = [value['loudness'] for value in d.values()]
        df['mode'] = [value['mode'] for value in d.values()]
        df['speechiness'] = [value['speechiness'] for value in d.values()]
        df['acousticness'] = [value['acousticness'] for value in d.values()]
        df['instrumentalness'] = [value['instrumentalness'] for value in d.values()]
        df['liveness'] = [value['liveness'] for value in d.values()]
        df['valence'] = [value['valence'] for value in d.values()]
        df['tempo'] = [value['tempo'] for value in d.values()]
        df['type'] = [value['type'] for value in d.values()]
        df['id'] = [value['id'] for value in d.values()]
        df['uri'] = [value['uri'] for value in d.values()]
        df['track_href'] = [value['track_href'] for value in d.values()]
        df['analysis_url'] = [value['analysis_url'] for value in d.values()]
        df['duration_ms'] = [value['duration_ms'] for value in d.values()]
        df['time_signature'] = [value['time_signature'] for value in d.values()]
```
